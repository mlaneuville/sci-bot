```python
import json
import os
import random
import subprocess
import time

from grobidclient import grobidclient
```

```python
def _parsepdf(path):
        service = 'processFulltextDocument'
        generateIDs = False
        consolidate_header = False
        consolidate_citations = False
        force = False
        teiCoordinates = False

        client = grobidclient(config_path='grobid-config.json')

        start_time = time.time()

        client.process(path, path, 20, service, generateIDs,
                       consolidate_header, consolidate_citations, force,
                       teiCoordinates)

        runtime = round(time.time() - start_time, 3)
        print(runtime)
```

```python
# stats 
stats = {'.json':0, '.pdf':0, '.xml':0}
for (dirpath, dirnames, fnames) in os.walk('pdf/'):
    for fname in fnames:
        stats[os.path.splitext(fname)[-1]] += 1
print(stats)
```

```python
# have json, try to get pdf 
json_no_pdf = []
for (dirpath, dirnames, fnames) in os.walk('pdf/'):
    for fname in fnames:
        if fname.endswith('.json'):
            if not os.path.isfile(os.path.join(dirpath, os.path.splitext(fname)[0] + '.pdf')):
                doi = os.path.join(dirpath, fname)[4:-5]
                json_no_pdf.append(doi)

print("%d papers with json and no pdf" % len(json_no_pdf))
```

```python
random.shuffle(json_no_pdf)
for doi in json_no_pdf:
    subprocess.run(['python3', 'sci-bot.py', '--doi', doi])
    time.sleep(5)
```

```python
# have pdf but no json
pdf_no_json = []
for (dirpath, dirnames, fnames) in os.walk('pdf/'):
    for fname in fnames:
        if fname.endswith('.pdf'):
            if not os.path.isfile(os.path.join(dirpath, os.path.splitext(fname)[0] + '.json')):
                doi = os.path.join(dirpath, fname)[4:-4]
                pdf_no_json.append(doi)

print("%d papers with pdf and no json" % len(pdf_no_json))
```

```python
for doi in pdf_no_json:
    subprocess.run(['python3', 'sci-bot.py', '--doi', doi])
    time.sleep(5)
```

```python
# have pdf but no xml
pdf_no_xml = []
for (dirpath, dirnames, fnames) in os.walk('pdf/'):
    for fname in fnames:
        if fname.endswith('.pdf'):
            if not os.path.isfile(os.path.join(dirpath, os.path.splitext(fname)[0] + '.tei.xml')):
                pdf_no_xml.append(dirpath)
                
print("%d papers with pdf and no xml" % len(pdf_no_xml))
pdf_no_xml = list(set(pdf_no_xml))
```

```python
for f in pdf_no_xml:
    _parsepdf(f)
    print("done")
    time.sleep(2)
```
